```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import re
import os
import unicodedata
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, f1_score, accuracy_score, confusion_matrix
from transformers import (
    AutoTokenizer, 
    TrainingArguments, 
    Trainer,
    DataCollatorWithPadding,
    RobertaPreTrainedModel, 
    RobertaModel, 
    AutoConfig
)
from datasets import Dataset, DatasetDict
from underthesea import word_tokenize



MODE = "VICOHATE_MTL_CLEAN"  
MODEL_CHECKPOINT = "vinai/phobert-base-v2" 


PHASE_2_REMOVE_STOPWORDS = False 

# ÄÆ°á»ng dáº«n file (Dá»¯ liá»‡u sáº¡ch v2)
TRAIN_PATH = "/kaggle/input/vicohatev2/train.csv"
VAL_PATH = "/kaggle/input/vicohatev2/val.csv"
TEST_PATH = "/kaggle/input/vicohatev2/test.csv"

OUTPUT_DIR = f"./results_vicohate_mtl_clean"

print(f"ğŸš€ ÄANG KHá»I Äá»˜NG: {MODE}")
print(f"ğŸ§¹ Cháº¿ Ä‘á»™ lÃ m sáº¡ch: Full Pipeline (Stopwords: {PHASE_2_REMOVE_STOPWORDS})")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

vietnamese_stopwords = set([
    "thÃ¬", "lÃ ", "mÃ ", "cá»§a", "nhÆ°", "nhá»¯ng", "cÃ¡c", "vÃ ", "hoáº·c", "ngÆ°á»i", 
    "cÃ¡i", "Ä‘Æ°á»£c", "táº¡i", "bá»‹", "bá»Ÿi", "vÃ¬", "do", "Ä‘á»ƒ", "khi", "trong", 
    "trÃªn", "dÆ°á»›i", "vá»›i", "ra", "vÃ o", "lÃªn", "xuá»‘ng"
])

teencode_dict = {
    "k": "khÃ´ng", "ko": "khÃ´ng", "kh": "khÃ´ng", "hok": "khÃ´ng", "hong": "khÃ´ng",
    "dc": "Ä‘Æ°á»£c", "Ä‘c": "Ä‘Æ°á»£c", "dk": "Ä‘Æ°á»£c",
    "t": "tao", "m": "mÃ y", "mk": "mÃ¬nh", "mn": "má»i ngÆ°á»i",
    "bth": "bÃ¬nh thÆ°á»ng", "bt": "bÃ¬nh thÆ°á»ng",
    "ntn": "nhÆ° tháº¿ nÃ o", "kq": "káº¿t quáº£",
    "trc": "trÆ°á»›c", "ng": "ngÆ°á»i", "n": "nÃ³",
    "ae": "anh em", "ad": "admin", "vs": "vá»›i", "wa": "quÃ¡",
    "uk": "á»«", "ukm": "á»«", "uh": "á»«", "thik": "thÃ­ch", "iu": "yÃªu",
    "dm": "Ä‘á»‹t máº¹", "Ä‘m": "Ä‘á»‹t máº¹", "dmm": "Ä‘á»‹t máº¹ mÃ y", "Ä‘km": "Ä‘á»‹t máº¹", 
    "vl": "vÃ£i lá»“n", "vcl": "vÃ£i lá»“n", "vkl": "vÃ£i lá»“n", "vlon": "vÃ£i lá»“n",
    "dell": "Ä‘Ã©o", "Ä‘Ã©o": "Ä‘Ã©o", "Ã©o": "Ä‘Ã©o", "dek": "Ä‘Ã©o",
    "cc": "cá»¥c cá»©t", "cl": "cÃ¡i lá»“n", "ml": "máº·t lá»“n",
    "xl": "xáº¡o lá»“n", "xÃ m l": "xÃ m lá»“n",
    "chet": "cháº¿t", "chjt": "chá»‹ch",
    "ngu": "ngu", "oc cho": "Ã³c chÃ³", "cho de": "chÃ³ Ä‘áº»",
    "bodo": "bÃ² Ä‘á»", "dlv": "dÆ° luáº­n viÃªn", 
    "cali": "pháº£n Ä‘á»™ng", "3que": "ba que", "3 que": "ba que",
    "namky": "nam ká»³", "backy": "báº¯c ká»³", "bake": "báº¯c ká»³", "bac ky": "báº¯c ká»³",
    "parky": "báº¯c ká»³", "paki": "báº¯c ká»³",
    "namcong": "nam cá»™ng", "vietcong": "viá»‡t cá»™ng", "vc": "viá»‡t cá»™ng",
    "cs": "cá»™ng sáº£n", "dcs": "Ä‘áº£ng cá»™ng sáº£n",
    "8 keo": "bÃ¡c há»“"
}

def advanced_clean_text(text):
    if not isinstance(text, str): return ""
    
    # 1. Normalize Unicode (Chuáº©n NFC: sá»­a lá»—i 'Ã²a' vs 'oÃ ')
    text = unicodedata.normalize('NFC', text)
    
    # 2. Lowercase (Chuyá»ƒn thÆ°á»ng)
    text = text.lower()
    
    # 3. Delete Links (XÃ³a URL)
    text = re.sub(r'http\S+|www\.\S+', '', text)
    
    # 4. Handle Icons/Emojis (XÃ³a icon - Regex dáº£i Unicode Emoji)
    text = re.sub(r'[\U00010000-\U0010ffff]', '', text) 
    
    # 5. Delete Redundant Characters (KÃ½ tá»± láº·p >2 láº§n -> rÃºt gá»n)
    # VÃ­ dá»¥: "nguuuuuu" -> "ngu", "vl....." -> "vl."
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 6. Normalize Teencode (DÃ¹ng tá»« Ä‘iá»ƒn Ä‘Ã£ Ä‘á»‹nh nghÄ©a)
    # Lá»c kÃ½ tá»± Ä‘áº·c biá»‡t trÆ°á»›c khi map teencode Ä‘á»ƒ trÃ¡nh lá»—i
    text = re.sub(r'[^\w\s]', ' ', text) 
    words = text.split()
    new_words = []
    for w in words:
        if w in teencode_dict:
            new_words.append(teencode_dict[w])
        else:
            # PHASE 2: Remove Stop Words
            if PHASE_2_REMOVE_STOPWORDS and w in vietnamese_stopwords:
                continue
            new_words.append(w)
            
    text = " ".join(new_words)
    
    # 7. Redundant Spaces (XÃ³a khoáº£ng tráº¯ng thá»«a)
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

def segment_text(text):
    if not isinstance(text, str) or not text.strip(): return ""
    try: return word_tokenize(text, format="text")
    except: return text

def smart_truncate(context):
    if not isinstance(context, str): return ""
    sentences = re.split(r'(?<=[.?!])\s+', context)
    if len(sentences) <= 2: return context
    return sentences[0] + " ... " + sentences[-1]

# ==============================================================================
# 2. CHUáº¨N Bá»Š Dá»® LIá»†U
# ==============================================================================
train_df = pd.read_csv(TRAIN_PATH)
val_df = pd.read_csv(VAL_PATH)
test_df = pd.read_csv(TEST_PATH)

# Class Weights
class_weights = np.array([1.0, 1.0, 2.0]) 
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(device)

tokenizer = AutoTokenizer.from_pretrained(MODEL_CHECKPOINT)

def preprocess_function(examples):
    # Ãp dá»¥ng hÃ m lÃ m sáº¡ch má»›i (advanced_clean_text)
    raw_contexts = [smart_truncate(str(c)) for c in examples["context_text"]]
    contexts = [advanced_clean_text(c) for c in raw_contexts]
    contexts_seg = [segment_text(c) for c in contexts]
    
    comments = [advanced_clean_text(str(c)) for c in examples["comment_text"]]
    comments_seg = [segment_text(c) for c in comments]
    
    return tokenizer(contexts_seg, comments_seg, truncation=True, max_length=256, padding=False)

raw_datasets = DatasetDict({
    "train": Dataset.from_pandas(train_df),
    "validation": Dataset.from_pandas(val_df),
    "test": Dataset.from_pandas(test_df),
})

tokenized_datasets = raw_datasets.map(preprocess_function, batched=True)


tokenized_datasets = tokenized_datasets.remove_columns(["comment_text", "context_text", "id"])
tokenized_datasets = tokenized_datasets.rename_column("final_label", "labels")
tokenized_datasets = tokenized_datasets.rename_column("stance_label_num", "labels_stance") 
tokenized_datasets.set_format("torch")
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)


class ViCoHateMultiTask(RobertaPreTrainedModel):
    def __init__(self, config):
        super().__init__(config)
        self.roberta = RobertaModel(config)
        self.classifier_hate = nn.Linear(config.hidden_size, 3)
        self.classifier_stance = nn.Linear(config.hidden_size, 3)
        self.init_weights()

    def forward(self, input_ids, attention_mask=None, token_type_ids=None, labels=None, labels_stance=None):
        outputs = self.roberta(input_ids, attention_mask=attention_mask, token_type_ids=token_type_ids)
        sequence_output = outputs[0][:, 0, :] 

        logits_hate = self.classifier_hate(sequence_output)
        logits_stance = self.classifier_stance(sequence_output)

        loss = None
        if labels is not None and labels_stance is not None:
            # Loss Hate
            loss_fct_hate = nn.CrossEntropyLoss(weight=class_weights_tensor.to(logits_hate.device))
            loss_hate = loss_fct_hate(logits_hate, labels)
            
            # Loss Stance
            loss_fct_stance = nn.CrossEntropyLoss()
            loss_stance = loss_fct_stance(logits_stance, labels_stance)
            
            loss = loss_hate + 0.5 * loss_stance

        output = (logits_hate,) + outputs[2:]
        return ((loss,) + output) if loss is not None else output


config = AutoConfig.from_pretrained(MODEL_CHECKPOINT, num_labels=3)
model = ViCoHateMultiTask.from_pretrained(MODEL_CHECKPOINT, config=config).to(device)

def compute_metrics(eval_preds):
    logits, labels = eval_preds
    if isinstance(logits, tuple): logits = logits[0]
    if isinstance(labels, tuple): labels = labels[0]
    
    predictions = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, predictions)
    f1_macro = f1_score(labels, predictions, average='macro')
    report = classification_report(labels, predictions, output_dict=True, zero_division=0)
    
    return {
        "accuracy": acc,
        "f1_macro": f1_macro,
        "f1_hate": report['2']['f1-score'],      
        "precision_hate": report['2']['precision'],
        "recall_hate": report['2']['recall']
    }

training_args = TrainingArguments(
    output_dir=OUTPUT_DIR,
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=7,
    weight_decay=0.01,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="f1_hate",
    save_total_limit=1,
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    processing_class=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)


print(f"\nğŸš€ Báº®T Äáº¦U HUáº¤N LUYá»†N: {MODE}...")
trainer.train()

save_path = f"./final_model_mtl_clean"
trainer.save_model(save_path)
tokenizer.save_pretrained(save_path)
print(f"ğŸ’¾ ÄÃ£ lÆ°u model táº¡i: {save_path}")

# ÄÃ¡nh giÃ¡
print("\nâ³ Äang Ä‘Ã¡nh giÃ¡ trÃªn táº­p Test...")
predictions = trainer.predict(tokenized_datasets["test"])
logits = predictions.predictions[0] if isinstance(predictions.predictions, tuple) else predictions.predictions
y_pred = np.argmax(logits, axis=-1)
y_true = predictions.label_ids
if isinstance(y_true, tuple): y_true = y_true[0]

print(f"\nğŸ“Š Káº¾T QUáº¢ CUá»I CÃ™NG ({MODE}):")
print(classification_report(y_true, y_pred, target_names=["Safe", "Offensive", "Hate"], digits=4))

# Váº½ Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
plt.figure(figsize=(8, 6))
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues', 
            xticklabels=["Safe", "Offensive", "Hate"], 
            yticklabels=["Safe", "Offensive", "Hate"])
plt.xlabel('Dá»± Ä‘oÃ¡n (Predicted)')
plt.ylabel('Thá»±c táº¿ (Ground Truth)')
plt.title(f'Confusion Matrix - {MODE}')
plt.show()

print("\nâœ… HOÃ€N Táº¤T!")
```
