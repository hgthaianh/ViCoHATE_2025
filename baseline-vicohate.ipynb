```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import re
import os
import unicodedata
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, f1_score, accuracy_score, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
from transformers import (
    AutoTokenizer, 
    AutoModelForSequenceClassification, 
    TrainingArguments, 
    Trainer,
    DataCollatorWithPadding,
    AutoConfig
)
from datasets import Dataset, DatasetDict
from underthesea import word_tokenize


MODE = "BASELINE_CLEAN"  
MODEL_CHECKPOINT = "vinai/phobert-base-v2" 

PHASE_2_REMOVE_STOPWORDS = False 

# ƒê∆∞·ªùng d·∫´n file (D·ªØ li·ªáu s·∫°ch v2)
TRAIN_PATH = "/kaggle/input/vicohatev2/train.csv"
VAL_PATH = "/kaggle/input/vicohatev2/val.csv"
TEST_PATH = "/kaggle/input/vicohatev2/test.csv"

OUTPUT_DIR = f"./results_baseline_clean"

print(f"üöÄ ƒêANG KH·ªûI ƒê·ªòNG: {MODE}")
print(f"üßπ Ch·∫ø ƒë·ªô l√†m s·∫°ch: Advanced (Stopwords: {PHASE_2_REMOVE_STOPWORDS})")
print(f"ü§ñ Model: {MODEL_CHECKPOINT}")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


vietnamese_stopwords = set([
    "th√¨", "l√†", "m√†", "c·ªßa", "nh∆∞", "nh·ªØng", "c√°c", "v√†", "ho·∫∑c", "ng∆∞·ªùi", 
    "c√°i", "ƒë∆∞·ª£c", "t·∫°i", "b·ªã", "b·ªüi", "v√¨", "do", "ƒë·ªÉ", "khi", "trong", 
    "tr√™n", "d∆∞·ªõi", "v·ªõi", "ra", "v√†o", "l√™n", "xu·ªëng"
])

teencode_dict = {
    "k": "kh√¥ng", "ko": "kh√¥ng", "kh": "kh√¥ng", "hok": "kh√¥ng", "hong": "kh√¥ng",
    "dc": "ƒë∆∞·ª£c", "ƒëc": "ƒë∆∞·ª£c", "dk": "ƒë∆∞·ª£c",
    "t": "tao", "m": "m√†y", "mk": "m√¨nh", "mn": "m·ªçi ng∆∞·ªùi",
    "bth": "b√¨nh th∆∞·ªùng", "bt": "b√¨nh th∆∞·ªùng",
    "ntn": "nh∆∞ th·∫ø n√†o", "kq": "k·∫øt qu·∫£",
    "trc": "tr∆∞·ªõc", "ng": "ng∆∞·ªùi", "n": "n√≥",
    "ae": "anh em", "ad": "admin", "vs": "v·ªõi", "wa": "qu√°",
    "uk": "·ª´", "ukm": "·ª´", "uh": "·ª´", "thik": "th√≠ch", "iu": "y√™u",
    "dm": "ƒë·ªãt m·∫π", "ƒëm": "ƒë·ªãt m·∫π", "dmm": "ƒë·ªãt m·∫π m√†y", "ƒëkm": "ƒë·ªãt m·∫π", 
    "vl": "v√£i l·ªìn", "vcl": "v√£i l·ªìn", "vkl": "v√£i l·ªìn", "vlon": "v√£i l·ªìn",
    "dell": "ƒë√©o", "ƒë√©o": "ƒë√©o", "√©o": "ƒë√©o", "dek": "ƒë√©o",
    "cc": "c·ª•c c·ª©t", "cl": "c√°i l·ªìn", "ml": "m·∫∑t l·ªìn",
    "xl": "x·∫°o l·ªìn", "x√†m l": "x√†m l·ªìn",
    "chet": "ch·∫øt", "chjt": "ch·ªãch",
    "ngu": "ngu", "oc cho": "√≥c ch√≥", "cho de": "ch√≥ ƒë·∫ª",
    "bodo": "b√≤ ƒë·ªè", "dlv": "d∆∞ lu·∫≠n vi√™n", 
    "cali": "ph·∫£n ƒë·ªông", "3que": "ba que", "3 que": "ba que",
    "namky": "nam k·ª≥", "backy": "b·∫Øc k·ª≥", "bake": "b·∫Øc k·ª≥", "bac ky": "b·∫Øc k·ª≥",
    "parky": "b·∫Øc k·ª≥", "paki": "b·∫Øc k·ª≥",
    "namcong": "nam c·ªông", "vietcong": "vi·ªát c·ªông", "vc": "vi·ªát c·ªông",
    "cs": "c·ªông s·∫£n", "dcs": "ƒë·∫£ng c·ªông s·∫£n",
    "8 keo": "b√°c h·ªì"
}

def advanced_clean_text(text):
    if not isinstance(text, str): return ""
    
    # 1. Normalize Unicode (Chu·∫©n NFC)
    text = unicodedata.normalize('NFC', text)
    
    # 2. Lowercase
    text = text.lower()
    
    # 3. Delete Links
    text = re.sub(r'http\S+|www\.\S+', '', text)
    
    # 4. Handle Icons/Emojis (Regex d·∫£i Unicode)
    text = re.sub(r'[\U00010000-\U0010ffff]', '', text) 
    
    # 5. Delete Redundant Characters (K√Ω t·ª± l·∫∑p >2 l·∫ßn)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # 6. Normalize Teencode & Stopwords
    text = re.sub(r'[^\w\s]', ' ', text)
    words = text.split()
    new_words = []
    for w in words:
        if w in teencode_dict:
            new_words.append(teencode_dict[w])
        else:
            if PHASE_2_REMOVE_STOPWORDS and w in vietnamese_stopwords:
                continue
            new_words.append(w)
            
    text = " ".join(new_words)
    
    # 7. Redundant Spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def segment_text(text):
    if not isinstance(text, str) or not text.strip(): return ""
    try: return word_tokenize(text, format="text")
    except: return text


train_df = pd.read_csv(TRAIN_PATH)
val_df = pd.read_csv(VAL_PATH)
test_df = pd.read_csv(TEST_PATH)

# T√≠nh Class Weights (Balanced cho Baseline)
labels = train_df['final_label'].values
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(labels), y=labels)
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float).to(device)
print(f"‚öñÔ∏è Class Weights (Balanced): {class_weights}")

tokenizer = AutoTokenizer.from_pretrained(MODEL_CHECKPOINT)

def preprocess_function(examples):
    # Ch·ªâ l·∫•y Comment (ƒê·∫∑c tr∆∞ng c·ªßa Baseline)
    comments = [advanced_clean_text(str(c)) for c in examples["comment_text"]]
    comments_seg = [segment_text(c) for c in comments]
    
    return tokenizer(
        comments_seg,
        truncation=True,
        max_length=256,
        padding=False
    )

raw_datasets = DatasetDict({
    "train": Dataset.from_pandas(train_df),
    "validation": Dataset.from_pandas(val_df),
    "test": Dataset.from_pandas(test_df),
})

tokenized_datasets = raw_datasets.map(preprocess_function, batched=True)
tokenized_datasets = tokenized_datasets.remove_columns(["comment_text", "context_text", "id", "stance_label_num"])
tokenized_datasets = tokenized_datasets.rename_column("final_label", "labels")
tokenized_datasets.set_format("torch")
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)


model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_CHECKPOINT, 
    num_labels=3,
    id2label={0: "Safe", 1: "Offensive", 2: "Hate"},
    label2id={"Safe": 0, "Offensive": 1, "Hate": 2}
).to(device)

class WeightedTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None):
        labels = inputs.get("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits")
        loss_fct = nn.CrossEntropyLoss(weight=class_weights_tensor)
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))
        return (loss, outputs) if return_outputs else loss

def compute_metrics(eval_preds):
    logits, labels = eval_preds
    predictions = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, predictions)
    f1_macro = f1_score(labels, predictions, average='macro')
    report = classification_report(labels, predictions, output_dict=True, zero_division=0)
    return {
        "accuracy": acc,
        "f1_macro": f1_macro,
        "f1_hate": report['2']['f1-score'],      
        "precision_hate": report['2']['precision'],
        "recall_hate": report['2']['recall']
    }

training_args = TrainingArguments(
    output_dir=OUTPUT_DIR,
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=7, 
    weight_decay=0.01,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="f1_hate",
    save_total_limit=1,
    report_to="none"
)

trainer = WeightedTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    processing_class=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

print(f"\nüöÄ B·∫ÆT ƒê·∫¶U HU·∫§N LUY·ªÜN: {MODE}...")
trainer.train()

# L∆∞u model
save_path = f"./final_model_{MODE.lower()}"
trainer.save_model(save_path)
tokenizer.save_pretrained(save_path)
print(f"üíæ ƒê√£ l∆∞u model t·∫°i: {save_path}")

# ƒê√°nh gi√° Test
print("\n‚è≥ ƒêang ƒë√°nh gi√° tr√™n t·∫≠p Test...")
predictions = trainer.predict(tokenized_datasets["test"])
y_pred = np.argmax(predictions.predictions, axis=-1)
y_true = predictions.label_ids

# In k·∫øt qu·∫£
print(f"\nüìä K·∫æT QU·∫¢ CU·ªêI C√ôNG ({MODE}):")
print(classification_report(y_true, y_pred, target_names=["Safe", "Offensive", "Hate"], digits=4))

# V·∫Ω Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
plt.figure(figsize=(8, 6))
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues', 
            xticklabels=["Safe", "Offensive", "Hate"], 
            yticklabels=["Safe", "Offensive", "Hate"])
plt.xlabel('D·ª± ƒëo√°n (Predicted)')
plt.ylabel('Th·ª±c t·∫ø (Ground Truth)')
plt.title(f'Confusion Matrix - {MODE}')
plt.show()

print("\n‚úÖ HO√ÄN T·∫§T!")
```
